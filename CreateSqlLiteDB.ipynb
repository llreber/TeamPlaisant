```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load
aid_data_to_load = "us_foreign_aid_modified.csv"

# Read Data File and store into Pandas Data Frames
aid_data = pd.read_csv(aid_data_to_load)
aid_data.head()
```

```python
from sqlalchemy import create_engine
engine = create_engine('sqlite:///aid_data.sqlite', echo=False)
```

```python
aid_data.to_sql("aid_data", con=engine, if_exists = "replace", chunksize = 10)
```

```python
engine.execute("SELECT * FROM aid_data WHERE fiscal_year = 2016").fetchall()
```
